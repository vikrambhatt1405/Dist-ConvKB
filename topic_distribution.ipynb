```python
import re
def topic_counts(file):
    topics={}
    with open(file, 'r',encoding="utf-8") as fin:
        for line in fin:
            if re.match("^/m/[a-zA-Z0-9_]*\t.*\t/m/[a-zA-Z0-9_]*",line):
                a=line.split("\t")
                key=a[1].split("/")[1]
                #print(key)
                topics[key] = topics.get(key, 0) + 1
                #print(topics[key])
    return topics
```

```python
topics=topic_counts("train.txt")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
sns.set_style("darkgrid")
%matplotlib inline
```

```python
import operator
sorted_topics = sorted(topics.items(), key=operator.itemgetter(1),reverse=True)
x=[]
y=[]
for (i,j) in sorted_topics:
    x.append(i)
    y.append(j)
```

```python
f=plt.figure(figsize=(15,15))
fig,(ax1,ax2)=plt.subplots(1,2,gridspec_kw={"width_ratios":[1, 2]})
fig.set_size_inches(14,14)
ax1=sns.barplot(x=x[:11],y=y[:11],ax=ax1)
labels=ax1.get_xticklabels()
ax1.set_xticklabels(labels,rotation=90)
sns.barplot(x=x[11:],y=y[11:],ax=ax2)
labels=ax2.get_xticklabels()
ax2.set_xticklabels(labels,rotation=90,fontdict={"fontsize":7})
fig.savefig("dist-1.pdf")
```
