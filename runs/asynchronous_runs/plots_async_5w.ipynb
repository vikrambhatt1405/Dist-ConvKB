```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
sns.set_style("darkgrid")
%matplotlib inline
```

```python
x_worker0 = []
y_worker0 = []

with open('run_worker0_.-tag-summary_Soft_Loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots,None)
    for row in plots:
        x_worker0.append(float(row[1]))
        y_worker0.append(float(row[2]))
x_worker0=np.array(x_worker0)
y_worker0=np.array(y_worker0)
```
