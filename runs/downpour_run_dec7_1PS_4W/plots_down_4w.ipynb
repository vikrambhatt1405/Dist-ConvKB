```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
sns.set_style("darkgrid")
%matplotlib inline
```

```python
x_worker0 = []
y_worker0 = []

with open('CSV_files/run_worker0_.-tag-summary_Soft_Loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots,None)
    for row in plots:
        x_worker0.append(float(row[1]))
        y_worker0.append(float(row[2]))
x_worker0=np.array(x_worker0)
y_worker0=np.array(y_worker0)
```

```python
fig=plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)
ax.plot(x_worker0[:600],y_worker0[:600],label='Worker0')
ax.plot(x_worker1[:600],y_worker1[:600],label='Worker1')
ax.plot(x_worker2[:600],y_worker2[:600],label='Worker2')
ax.plot(x_worker3[:600],y_worker3[:600],label='Worker3')

ax.set_xlabel("Global steps")
ax.set_ylabel("Loss")
ax.legend()
ax.annotate('',
            xy=(x_worker0[200],y_worker0[200]),  # theta, radius
            xytext=(0.7, 0.45),    # fraction, fraction
            textcoords='figure fraction',
            arrowprops=dict(facecolor='black', shrink=0.05,width=0.3),
            horizontalalignment='right',
            verticalalignment='top' )

ax.set_title("Learning cure for DOWNPOUR with 1 Parameter server and 4 Workers(FB15K-2137)")


a = plt.axes([.6, .4, .2, .2],frame_on=True,clip_on=False)
a.plot(x_worker0[580:600],y_worker0[580:600],label='worker 0')
a.plot(x_worker1[580:600],y_worker1[580:600],label='worker 1')
a.plot(x_worker2[580:600],y_worker2[580:600],label='worker 2')
a.plot(x_worker3[580:600],y_worker3[580:600],label='worker 3')
a.set_title('Illustration of racing condition')
a.legend()
plt.xticks([])
plt.savefig("Down_4w.pdf")
```
