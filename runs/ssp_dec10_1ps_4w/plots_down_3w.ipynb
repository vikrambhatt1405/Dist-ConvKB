```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
sns.set_style("darkgrid")
%matplotlib inline
```

```python
x_worker0 = []
y_worker0 = []

with open('run_worker0_.-tag-summary_Soft_Loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots,None)
    for row in plots:
        x_worker0.append(float(row[1]))
        y_worker0.append(float(row[2]))
x_worker0=np.array(x_worker0)
y_worker0=np.array(y_worker0)
```

```python
from scipy.signal import savgol_filter
fig=plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)

ax.plot(x_worker0,savgol_filter(y_worker0, 101, 2),label='Worker0')
ax.plot(x_worker1,savgol_filter(y_worker1, 101, 2),label='Worker1')
ax.plot(x_worker2,savgol_filter(y_worker2, 101, 2),label='Worker2')
ax.plot(x_worker3,savgol_filter(y_worker3, 101, 2),label='Worker2')
ax.set_xlabel("Global steps")
ax.set_ylabel("Loss")
ax.legend()

ax.set_title("Learning cure for DOWNPOUR with 1 Parameter server and 3 Workers(FB15K-2137)")

plt.savefig("stale_4w.pdf")
```
