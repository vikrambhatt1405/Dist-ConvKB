```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
sns.set_style("darkgrid")
%matplotlib inline
```

```python
x_worker0 = []
y_worker0 = []

with open('run_worker0_.-tag-summary_Soft_Loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots,None)
    for row in plots:
        x_worker0.append(float(row[1]))
        y_worker0.append(float(row[2]))
x_worker0=np.array(x_worker0)
y_worker0=np.array(y_worker0)
```

```python
x_worker1 = []
y_worker1 = []

with open('run_worker1_.-tag-summary_Soft_Loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots,None)
    for row in plots:
        x_worker1.append(float(row[1]))
        y_worker1.append(float(row[2]))
x_worker1=np.array(x_worker1)
y_worker1=np.array(y_worker1)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharex=True)
ax1.plot(x_worker0, y_worker0)
ax1.set_title('Worker 0')

ax2.scatter(x_worker1, y_worker1)
ax2.set_xticklabels([])
ax2.set_title('Worker 1')
```

```python
plt.plot(x_worker0[:600],y_worker0[:600],label='Worker0')
plt.plot(x_worker1[:600],y_worker1[:600],label='Worker1')
plt.xlabel("Global steps")
plt.ylabel("Loss")
plt.legend()
plt.title("Synchronous training with 2 workers and 1 parameter server")
plt.savefig("sync_2w.pdf")
```
