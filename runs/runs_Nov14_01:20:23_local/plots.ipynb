```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
sns.set_style("darkgrid")
%matplotlib inline
```

```python
x_worker0 = []
y_worker0 = []

with open('run_.-tag-summary_Soft_Loss.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots,None)
    for row in plots:
        x_worker0.append(float(row[1]))
        y_worker0.append(float(row[2]))
x_worker0=np.array(x_worker0)
y_worker0=np.array(y_worker0)
```

```python
plt.plot(x_worker0[:600],y_worker0[:600],label='Worker0')
plt.xlabel("Global steps")
plt.ylabel("Loss")
plt.legend()
plt.title("Training on local machine Intel® Core™ i7-7700K CPU @ 4.20GHz")
plt.savefig("local.pdf")
```
